```python
from tensorflow.layers import conv2d, max_pooling2d, conv2d_transpose
import tensorflow as tf
import tensorflow.contrib.slim as slim

class Autoencoder:
    def build_model(self):
        self.input = tf.placeholder(dtype=tf.float32, shape=(None, IMAGE_HEIGHT, IMAGE_WEIGHT, 3))

        x = conv2d(self.input, filters=128, kernel_size=(3, 3), activation=tf.nn.relu, padding='SAME')
        x = max_pooling2d(x, pool_size=(2, 2), strides=(1, 1), padding='SAME')
        
        x = conv2d(x, filters=64, kernel_size=(3, 3), activation=tf.nn.relu, padding='SAME')
        x = max_pooling2d(x, pool_size=(2, 2), strides=(1, 1), padding='SAME')

        x = conv2d(x, filters=64, kernel_size=(3, 3), activation=tf.nn.relu, padding='SAME')
        x = max_pooling2d(x, pool_size=(2, 2), strides=(1, 1), padding='SAME')

        x = conv2d(x, filters=32, kernel_size=(3, 3), activation=tf.nn.relu, padding='SAME')
        x = max_pooling2d(x, pool_size=(2, 2), strides=(1, 1), padding='SAME')


        x = conv2d(x, filters=32, kernel_size=(3, 3), activation=tf.nn.relu, padding='SAME')
        encoded = max_pooling2d(x, pool_size=(2, 2), strides=(1, 1), padding='SAME')

        # Decoder
        x = conv2d(encoded, filters=32, kernel_size=(3, 3), activation=tf.nn.relu, padding='SAME')
        x = conv2d_transpose(encoded, filters=32, kernel_size=(2, 2))


        x = conv2d(x, filters=64, kernel_size=(3, 3), activation=tf.nn.relu, padding='SAME')
        x = conv2d_transpose(x, filters=64, kernel_size=(3, 3), strides=(1, 1))

        x = conv2d(x, filters=64, kernel_size=(3, 3), activation=tf.nn.relu)
        x = conv2d_transpose(x, filters=64, kernel_size=(2, 2))

        x = conv2d(x, filters=128, kernel_size=(3, 3), activation=tf.nn.relu)
        x = conv2d_transpose(x, filters=128, kernel_size=(2, 2))


        x = conv2d(x, filters=128, kernel_size=(3, 3), activation=tf.nn.relu)
        self.output = conv2d_transpose(x, filters=3, kernel_size=(2, 2), strides=(1, 1))        
        
    def summary(self):
        model_vars = tf.trainable_variables()
        slim.model_analyzer.analyze_vars(model_vars, print_info=True)
```

```python
IMAGE_HEIGHT = 100
IMAGE_WEIGHT = 100

# tf.reset_default_graph() 
autoencoder = Autoencoder()
autoencoder.build_model()
autoencoder.summary()

print(autoencoder.input)
print(autoencoder.output)
```
