```python
import cv2
import random

class ImageGenerator:
    def __init__(self, image_height, image_width):
        self.image_height = image_height
        self.image_width = image_width
        
    def get_X_image(self, path, is_augment = False):
        image = cv2.imread(path)
        
        if is_augment and np.random.rand() < 0.6:
            image = self.augment(image)
            
        image = self.preprocess(image)

        return image
        

    def preprocess(self, image):
        image = self.resize(image)
        image = self.normalize(image)
        return image
    
    def augment(self, image):
        return image

    def normalize(self, image):
        return cv2.normalize(image,  image, 0, 1, cv2.NORM_MINMAX,  dtype=cv2.CV_32F)
    
    def resize(self, image):
        return cv2.resize(image, (self.image_width, self.image_height), cv2.INTER_AREA)
    
    def augment(self, image):
        image = self.random_flip(image)
        image = self.random_rotate(image)
        return image
    
    def random_flip(self, image):
        if np.random.rand() < 0.5:
            image = cv2.flip(image, 1)
        return image
    
    def random_rotate(self, image):
        max_angle = 15
        angle = random.randint(-max_angle,max_angle)
        
        image_center = tuple(np.array(image.shape[1::-1]) / 2)
        rot_mat = cv2.getRotationMatrix2D(image_center, angle, 1.0)
        result = cv2.warpAffine(image, rot_mat, image.shape[1::-1], flags=cv2.INTER_LINEAR)
        return result
```
