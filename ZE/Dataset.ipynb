```python
class Dataset:
    def __init__(self, X_path, Y_path, batch_size, image_generator, is_augment):
        self.X_path = X_path
        self.Y_path = Y_path
        self.batch_size = batch_size
        self.x_names = os.listdir(self.X_path)

        self.y_names = os.listdir(self.Y_path)

        self.image_generator = image_generator
        self.is_augment = is_augment
#         for mac users, remove extra file name
        while '.DS_Store' in self.x_names: self.x_names.remove('.DS_Store')
            
            
        #PRIVATE
        self.batches_count = int(np.floor(len(self.x_names) / self.batch_size))
        self.reset_index()
        
        
    def reset_index(self, shuffle = True):
        self.current_batch_index = 0
        
        if shuffle:
            np.random.shuffle(self.x_names)
            
    def get_next_batch(self, batch_size):
        batch_X = [batch_size]
        batch_Y = [batch_size]
        
        start_slice_index = self.current_batch_index * batch_size
        end_slice_index = (self.current_batch_index + 1) * batch_size

        
        batch_X_names = self.x_names[start_slice_index:end_slice_index]
        #batch_Y_names = self.y_names[start_slice_index:end_slice_index]
        
        images = self.get_image_Xs(batch_X_names)
        
        if self.current_batch_index < self.batches_count - 1:
            self.current_batch_index += 1

        return images, images 


    def get_image_Xs(self, batch_X_names):
        images = []
        for index, name in enumerate(batch_X_names):
            image_path = self.X_path + '/' + name
            image = image_generator.get_X_image(image_path, self.is_augment)
            images.append(image)
            
        return images
```
